```python
from ultralytics import YOLO

model = YOLO("yolo11n-cls.pt")

results = model.train(data='/content/car', epochs=30, imgsz=640)
```

```python
load_model = YOLO('/content/runs/classify/train/weights/best.pt')

results = model('/content/02-how-to-remove-small-car-dents.webp')
results[0].show()
```

```python
import numpy as np
for result in results:
    probs = result.probs.data.cpu().numpy()
    class_names = {0: "dent", 1: "glass shatter", 2: "lamp broken", 3: "scratch"}

    for class_id, class_name in class_names.items():
        confidence = probs[class_id]
        if round(confidence,2) > 0.5:  # Strong classification confidence
            print(f"🔴 **{class_name.upper()} DETECTED** — The model classifies this as {confidence:.2%} likely.\n")
        elif round(confidence,2) > 0.1:  # Weak classification confidence
            print(f"🟡 **Possible {class_name.upper()}** — The model suggests {confidence:.2%} likelihood.\n")
        else:  # No damage classified
            print(f"🟢 **No {class_name.upper()} detected.**\n")
```
